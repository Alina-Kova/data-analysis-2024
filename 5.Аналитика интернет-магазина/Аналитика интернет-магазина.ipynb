## Приоритизация гипотез.

```python
# считываем данные из .csv файла с гипотезами и переносим их в датафрейм
data = pd.read_csv('https://code.s3.yandex.net/datasets/hypothesis.csv')
```

Перед нами 9 гипотез по увеличению выручки интернет-магазина с указанными параметрами Reach, Impact, Confidence, Effort.  

Описание данных датасета `data`:  

`Hypothesis` — краткое описание гипотезы;  
`Reach` — охват пользователей по 10-балльной шкале;  
`Impact` — влияние на пользователей по 10-балльной шкале;  
`Confidence` — уверенность в гипотезе по 10-балльной шкале;  
`Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение `Efforts`, тем дороже проверка гипотезы.  

Гипотезы, представленные к рассмотрению:
- *Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей*
- *Запустить собственную службу доставки, что сократит срок доставки заказов*	 
- *Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа*  - *Изменить структура категорий, что увеличит конверсию, т.к. пользователи быстрее найдут нужный товар*  
- *Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей*  
- *Добавить страницу отзывов клиентов о магазине, что позволит увеличить количество заказов*  
- *Показать на главной странице баннеры с актуальными акциями и распродажами, чтобы увеличить конверсию*
- *Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок*
- *Запустить акцию, дающую скидку на товар в день рождения*

Для того, чтобы определить, какая из гипотез будет приоритетной, можем применить методы расчета приоритезации - ICE и RICE, основываясь на значениях, указанных в параметрах Reach, Impact, Confidence, Effort.

### Применение фреймворка ICE

Применим **фреймворк ICE** для приоритизации гипотез и отсортируем их по убыванию приоритета.  
Для расчета приоритизации задач методом ICE умножим `Impact` (уровень вляния изменений на пользователей, их опыт и удовлетворение от продукта) на `Confidence` (уровень уверенности того, что изменение повлияет на пользователя тем или иным образом) и разделим результат умножения на `Efforts` (стоимость тестирования гипотезы).

```python
#расчитываем приоритезацию гипотез по методу ICE, создав для резултата расчета отдельную колонку ICE
data['ice'] = data['impact'] * data['confidence'] / data['efforts']
#округляем результат до 2 цифр после запятой
data['ice'] = data['ice'].round(2)
#выводим результат на экран и сортируем по убыванию колонки ICE
display(data[['hypothesis', 'ice']].sort_values(by='ice', ascending=False))
```

### Применение фреймворка RICE

Применим **фреймворк RICE** для приоритизации гипотез и отсортируем их по убыванию приоритета.  
Для расчета приоритизации задач методом RICE умножим `Reach` (количество пользователей, которых затронет то или иное изменение) на `Impact` (уровень вляния изменений на пользователей, их опыт и удовлетворение от продукта), затем умножим на `Confidence` (уровень уверенности того, что изменение повлияет на пользователя тем или иным образом) и разделим результат умножения на `Efforts` (стоимость тестирования гипотезы).

```python
#расчитываем приоритезацию гипотез по методу RICE, создав для резултата расчета отдельную колонку RICE
data['rice'] = data['reach'] * data['impact'] * data['confidence'] / data['efforts']
#округляем результат до 2 цифр после запятой
data['rice'] = data['rice'].round(2)
#выводим результат на экран и сортируем по убыванию колонки RICE
display(data[['hypothesis', 'rice']].sort_values(by='rice', ascending=False))
```

**Вывод**  

При применении фреймформка ICE приоритетной гипотезой стала №8 - ***Запустить акцию, дающую скидку на товар в день рождения***. При применении RICE та же гипотеза упала на 5 место, а на первое c третьего места поднялась гипотеза №7 - ***Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок***. В целом порядок практически всех гипотез поменялся после расчета по методу RICE. Это связано с тем, что RICE учитывает количество пользователей, которых затронет изменение, которое предполагается внести (`Reach`). У гипотезы №7 самый высокий показатель `Reach`, поэтому именно она встала на первое место после применения расчета по RICH.  

При применении фреймфорка ICE главным фактором, оказывающим влиение на приоритизацию, был показатель `Impact` (сила влияния гипотезы) - чем показатель выше, тем приоритетнее гипотеза, но показатель `Reach` не учитывается.

**Исходя из расчетов, приоритетными можем считать гипотезы №7, 0, 6**, так как они находятся в топе списков при расчетах. Гипотезы, которые точно не приоритетны - № 4, 5, 3, 1.

### График кумулятивной выручки по группам.

Построим график кумулятивной выручки по группам и проанализируем его.  
Но, прежде чем построить график, для начала нам нужно собрать кумулятивные данные. Для этого создадим датафрейм `cumulativeData`, в котором будут столбцы:  
`date` — дата  
`group` — группа A/B-теста (A или B)  
`orders` — кумулятивное количество заказов на указанную дату в указанной группе  
`buyers` — кумулятивное количество пользователей, совершивших хотя бы один заказ, на указанную дату в указанной группе  
`revenue` — кумулятивная выручка на указанную дату в указанной группе (средний чек)  
`visitors` — кумулятивное количество посетителей интернет-магазина на указанную дату в определённой группе. 

```python
# создаем массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# получаем агрегированные кумулятивные по дням данные о заказах 
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], 
                                                                     orders['group'] == x['group'])]
                                     .agg({'date' : 'max', 
                                           'group' : 'max', 
                                           'transactionId' : 'nunique', 
                                           'visitorId' : 'nunique', 
                                           'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

# получаем агрегированные кумулятивные по дням данные о посетителях интернет-магазина 
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], 
                                                                         visitors['group'] == x['group'])]
                                       .agg({'date' : 'max', 
                                             'group' : 'max', 
                                             'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulativeData = ordersAggregated.merge(visitorsAggregated, 
                                        left_on=['date', 'group'], 
                                        right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

display(cumulativeData.head(5)) 
```

Проверим корректность созданного датафрейма `cumulativeData`, для этого взглянем на даты начала и окончания теста. Эти даты должны совпадать с теми, которые мы вычеслили для датасетов `orders` и `visitors` на этапе предобработки - 2019-08-01 - 2019-08-31.

```python
# проверим временной интервал, в который проводитлся тест
print('Дата начала теста:', cumulativeData['date'].min())
print('Дата окончания теста:', cumulativeData['date'].max())
```

Теперь можем построить графики кумулятивной выручки по группам.

```python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

# задаем размер графика
plt.figure(figsize=(15,7))
# задаем название графика
plt.title('Графики кумулятивной выручки по группам A/B-тестирования' + '\n', fontsize = 16, weight='bold')
# задаем названия осей и размер шрифта
plt.xlabel('\n' + 'Дата', fontsize = 16)
plt.ylabel('Сумма выручки, $' + '\n', fontsize = 16)

# Строим график выручки группы А
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='Группа A')

# Строим график выручки группы B
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='Группа B')

# задаем параметры делений на осях, устанавливаем сетку и задаем параметры легенды 
plt.tick_params(labelsize='large')
plt.grid(True)
plt.legend(fontsize='x-large');
```

**Вывод**  

Выручка ***стабильно растет как для группы А, так и для группы B*** каждый день в течение всего периода, что является хорошим покатателем. Но ***у группы B в даты 18-08-2019 и 19-08-2019 наблюдается резкий скачок выручки***. Это может означать, что среди пользователей группы А появился резкий всплеск числа заказов, либо были сделаны очень дорогие заказы. При этим после скачка выручки у группы В между группами образовался значительный разрыв вплоть до окончания теста. К выбросам мы вернемся и проанализируем их на следующих этапах исследования.

### Статистическая значимость различий в среднем числе заказов по «сырым» данным.

Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным, то есть без удаления аномальных пользователей.

Перед расчетом подготовим данные: cоберем общую таблицу, в которую войдут данные по ежедневным визитам пользователей.  

```python
# создаем таблицу с количеством пользователей по датам в группе А
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']


# создаем таблицу с количеством пользователей по датам в группе В
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']


# объединяем все созданные таблицы в единую
data = (
    visitorsADaily.merge(visitorsBDaily, left_on='date', right_on='date', how='left')
)

display(data.head(5)) 
```

Состав таблицы:  

`date` — дата  
`visitorsPerDateA` — количество пользователей в выбранную дату в группе A  
`visitorsPerDateB` — количество пользователей в выбранную дату в группе B

Теперь можем посчитать статистическую значимость различия в среднем количестве заказов между группами. 

Сформулируем гипотезы:  
**H0**: у групп A и B равное среднее количество заказов.  
**H1**: различия в среднем количестве заказов между группами A и B есть.

```python
# cоздадим переменные ordersByUsersA и ordersByUsersB со столбцами ['visitorsId', 'orders'] 
# и в них укажем число совершённых заказов для пользователей, которые заказывали хотя бы 1 раз
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['visitorId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['visitorId', 'orders']

# объявим переменные sampleA и sampleB, в которых пользователям из разных групп будет соответствовать 
# количество заказов, а тем, кто ничего не заказал, будут соответствовать нули
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
#определяем уровень статистической значимости 5%
alpha = .05 

# применим U-критерий Манна-Уитни
results = stats.mannwhitneyu(sampleA, sampleB)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: различия в среднем количестве заказов между группами A и B есть.')
else:
    print('Не получилось отвергнуть нулевую гипотезу: ' 
          'у групп A и B равное среднее количество заказов')


print()
# выводим результат статистической значимости по критерию Манна-Уитни
print('P-value равно', "{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))

# выводим относительный прирост среднего числа заказов группы B и округлим до трёх знаков после запятой
print('Показатель относительного прироста среднего числа заказов группы B к группе A по "сырым" данным = ', 
      "{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

**Вывод**  

P-value меньше уровня статистической значимости (5%) и составляет 1.1%. Это означает, что резльтат статистически значим, т.е. разница в среднем количестве заказов у групп А и В неслучайна. Соответственно, нулевую гипотезу отвеграем, подтверждая альтернативную: **различия в среднем количестве заказов между группами A и B есть**.  

При этом группа В выигрывает в приросте числа заказов относительно группы А на 16%. 


### Статистическая значимости различий в среднем чеке заказа  по «сырым» данным.

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным.  

Сформулируем гипотезы:  
**H0**: средний чек у групп A и B равен.  
**H1**: различия в среднем чеке между группами A и B есть.

```python
#определяем уровень статистической значимости 5%
alpha = .05 

# применим U-критерий Манна-Уитни
#передадим критерию mannwhitneyu() данные из колонки revenue - выручка с заказов
results = stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], 
                                          orders[orders['group']=='B']['revenue'])

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: различия в среднем чеке между группами A и B есть.')
else:
    print('Не получилось отвергнуть нулевую гипотезу: ' 
          'средний чек у групп A и B равен.')


print()
# выводим результат статистической значимости по критерию Манна-Уитни
print('P-value равно', '{0:.3f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], 
                                          orders[orders['group']=='B']['revenue'])[1]))

# выведем относительные различия в среднем чеке между группами A и B
print('Показатель относительного прироста среднего чека группы B к группе A по "сырым" данным равен',
      '{0:.3f}'.format(orders[orders['group']=='B']['revenue']
                       .mean()/orders[orders['group']=='A']['revenue'].mean()-1))
```

**Вывод**  

P-значение равно 82.9%, очевидно, превышая в разы уровень статистической значимости, который равен 5%. Это означает, что наблюдаемая разница в среднем чеке у групп А и В случайна. Нулевая гипотеза о равенстве подтверждается: **средний чек у групп A и B равен.**  

При этом средний чек группы В выше среднего чека группы А 28.7%. И посколько различие среднего чека между сегментами равно ~28%, то наблюдаемые различия не будут приняты ко вниманию при таком уровне статистической значимости, а, следовательно, были получены случайно.

### Статистическая значимость различий в среднем числе заказов по «очищенным» данным.


Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным.  

На этапах ранее мы определили, что подавляющее большинство пользователей совершает не более 1 заказа на сумму не более 26 785\\$, а показатели выше этих будем считать на аномалии. Теперь нам необходимо избавиться от аномальных значений и проверить гипотезы на "очищенных" данных. Таким образом мы уберем из датасета до 5% пользователей с заказами, превышающими стоимость 26 785 и до 1% c 2 и более заказами.

```python
# зададим переменные с указанием границ между данными, которые остаются в исследовательском анализе, и 
# аномальными значениями, которые выбрасываем из дальнейшего исследования
orders_number_limit = 1
revenue_limit = 26785
```

Сделаем срезы пользователей с числом заказов больше 1 — usersWithManyOrders и пользователей, совершивших заказы дороже 26 785 — usersWithExpensiveOrders. Объединим их в таблице abnormalUsers. 
Узнаем, сколько всего аномальных пользователей атрибутом shape.

```python
# даелаем срез пользователей с числом заказов от 2 и более
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > orders_number_limit]['visitorId'],
        ordersByUsersB[ordersByUsersB['orders'] > orders_number_limit]['visitorId'],
    ],
    axis=0,
)


# даелаем срез пользователей, совершивших заказы более, чем на 26 785
usersWithExpensiveOrders = orders[orders['revenue'] > revenue_limit]['visitorId']

# объединим два среза в таблице abnormalUsers 
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
# выводим первые 5 строк таблицы
print(abnormalUsers.head(5))
# считаем количество аномальных пользователей
print(abnormalUsers.shape[0])
```

**Всего 86 аномальных пользователя**. Теперь можем посчитать статистическую значимость в количестве заказов между группами А и В по "очищенным" данным и заодно узнаем, как 86 аномальных пользователя повлияли на результаты теста. 

```python
# подготовим выборки количества заказов по пользователям по группам А и В, исключая аномальных пользователей
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

Сформулируем гипотезы:  

**H0**: у групп A и B равное среднее количество заказов по очищенным данным.  
**H1**: различия в среднем количестве заказов по очищенным данным между группами A и B есть.

```python
#определяем уровень статистической значимости 5%
alpha = .05 

# применим U-критерий Манна-Уитни
#передадим критерию mannwhitneyu() данные из выборок
results = stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу:' + '\n' + 'различия в среднем количестве заказов по "очищенным" данным '
          'между группами A и B есть.')
else:
    print('Не получилось отвергнуть нулевую гипотезу: ' 
          'у групп A и B равное среднее количество заказов по "очищенным" данным.')

print()
# выводим результат статистической значимости по критерию Манна-Уитни    
print('P-value равно', '{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
# выводим относительный прирост среднего числа заказов группы B по очищенным данным 
# и округлим до трёх знаков после запятой
print('Показатель относительного прироста среднего числа заказов группы B к группе A по "очищенным" данным равен', 
      '{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```

**Вывод**  

P-value меньше уровня статистической значимости (5%) и составляет 0.9%. Это означает, что резльтат статистически значим, т.е. разница в среднем количестве заказов у групп А и В по 'очищенным' данным неслучайна. Соответственно, нулевую гипотезу отвеграем, подтверждая альтернативную: **различия в среднем количестве заказов по "очищенным" данным между группами A и B есть**.  

При этом группа В выигрывает в приросте числа заказов относительно группы А на 18.9%. Мы получили практически такой же результат, как и по "сырым" данным.

### Статистическая значимость различий в среднем чеке заказа по «очищенным» данным.

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным.

Сформулируем гипотезы:  
**H0**: средний чек по "очищенным" данным у групп A и B равен.  
**H1**: различия в среднем чеке по "очищенным" данным между группами A и B есть.

```python
#определяем уровень статистической значимости 5%
alpha = .05 

# применим U-критерий Манна-Уитни
#передадим критерию mannwhitneyu() данные из колонки revenue - выручка с заказов, 
# но исключая аномальных пользователей
results = stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: различия в среднем чеке по "очищенным" данным между группами A и B есть.')
else:
    print('Не получилось отвергнуть нулевую гипотезу: ' 
          'средний чек по "очищенным" данным у групп A и B равен.')

print()
# выводим результат статистической значимости по критерию Манна-Уитни
print('P-value равно', 
    '{0:.3f}'.format(
        stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
)

# выведем относительные различия в среднем чеке по "очищенным" между группами A и B
print('Показатель относительного прироста среднего чека по "очищенным" данным'
      ' группы B к группе A по "очищенным" данным равен',
    "{0:.3f}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
)
```

**Вывод**  

P-значение равно 76.2% и превышает в разы уровень статистической значимости, который равен 5%. Это означает, что наблюдаемая разница в среднем чеке по "очищенным" данным у групп А и В случайна. Нулевая гипотеза о равенстве подтверждается: **средний чек по "очищенным" данным у групп A и B равен**.  

При этом средний чек группы В проигрывает среднему чеку группы А на 2.2%.
