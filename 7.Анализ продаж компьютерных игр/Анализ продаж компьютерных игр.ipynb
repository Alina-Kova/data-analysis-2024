# Проект по анализу данных о продажах компьютерных игр  

 
Для проведения исследовательского анализа мы получили исторические данные о продажах компьютерных игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Данные предоставлены за период с 1980 до 2016 года включително. 

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию.

Система ESRB:  
`EC` - Для детей младшего возраста. Подходит для детей от 3 лет, неприемлемое содержание отсутствует.  
`E`	- Подходит для всех возрастных категорий. Игра не должна содержать звуков или изображений, которые могут напугать маленьких детей. Не должна использоваться ненормативная лексика.  
`E 10+`	- Подходит для лиц старше 10 лет. Игра может содержать умеренные формы насилия, а некоторые сцены могут пугать детей.  
`T`	- Подходит для лиц старше 13 лет. Игра может содержать более реалистичные и натуралистические сцены насилия.  
`M`	- Подходит для лиц старше 17 лет. Этот рейтинг используется, когда насилие становится реалистичным и похожим на реальную жизнь. Также может присутствовать ненормативная лексика, употребление табака, алкоголя или запрещенных наркотиков.  
`AO` - Подходит для взрослых старше 18 лет. Классификация «для взрослых» используется, когда в игре присутствует крайний уровень насилия и немотивированного убийства. Также может присутствовать романтизация наркотиков, азартных игр и сексуальной активности.  
`RP` - Категория ещё не присвоена. Играм с категорией RP еще не присвоен окончательный рейтинг ESRB.  
`K-A` - «Kids to Adults» , подходит для детей и взрослых. Содержание игры вполне подходит для возрастной категории от 6 лет и старше. Такие игры подходят для всех возрастов. Они могут содержать минимальное насилие или немного грубый разговорный язык. Этот рейтинг был заменён на «Everyone» 1 января 1998 года и может встречаться только в играх, изданных ранее, кроме The Simpsons Cartoon Studio.

В колонке наблюдаются выбивающиеся зачения - *tbd* и *nan*. *tbd* означает с английского  *to be determined* - еще не определено, поэтому можем предположить, что разработчики выпустили игры только в 2016г и оценка пользователей на текущий момент у игр отсутствует. ### Анализ платформ по количеству продаж

Проведем анализ лидирующих по продажам платформ. Установим, какие из них растут, а какие падают. Выберем несколько потенциально прибыльных платформ.

```python
#построим сводную таблицу с суммарными продажами по платформам
actual_game_data_pivot = actual_game_data.pivot_table(index='platform', 
                             values='total_sales', 
                             aggfunc='sum').sort_values(by='total_sales', ascending=False)

actual_game_data_pivot
```

```python
#построим цикл создания гистограммы для каждой платформы по количеству выпущенных игр с 2014 по 2016 года

# создаем список с интересующими нас годами
actual_years = [2014, 2015, 2016, 2017]

for i in range(0, len(actual_game_data_pivot.index)):
    actual_game_data[actual_game_data['platform'] == actual_game_data_pivot.index[i]].plot(kind='hist',
                                                                                           grid=True,
                                                                                           bins=actual_years,
                                                                                           histtype='step',
                                                                                           linewidth=5,
                                                                                           alpha=0.7,
                                                                                           legend=False)

    plt.title(actual_game_data_pivot.index[i], fontweight="bold")
    plt.xticks(actual_years)
    plt.xlabel('Год')
    plt.ylabel('Количество игр')
    plt.grid(True)
    plt.show()
```

### Анализ влияния отзывов на популярность платформ

Проведем анализ, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Для этого построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. Затем соотнесем выводы с продажами игр на других платформах.  

Для анализа возьмем платформу PS4, так как в предыдущем исследовании мы выяснили, что это самая популярная платформа с самыми высокими продажами игр.

```python
#сделаем два среза данных, убрав временные замены данных в оценках пользователей и критиков на 0, 
#а также оставив только платформу PS4
actual_user_score_ps4 = actual_game_data.query('(user_score > 0.0) and (platform == "PS4")')
actual_critic_score_ps4 = actual_game_data.query('(critic_score > 0.0) and (platform == "PS4")')

# построим диаграммы рассеяния с линией регрессии, чтобы посмотреть на корреляцию 
# количества продаж игр и оценок пользователей/критиков

fig = plt.figure(figsize=(15, 5))

ax1 = fig.add_subplot(121)
ax1 = sns.regplot(x=actual_critic_score_ps4["critic_score"], 
            y=actual_critic_score_ps4["total_sales"], 
            line_kws={"color":"r","alpha":0.7,"lw":5})
ax1.set_title('Влияние оценок критиков на продажи платформы PlayStation 4', size='medium', weight='bold')
ax1.set(xlabel='Оценка критиков', ylabel='Количество глобальных продаж (млн копий)')


ax2 = fig.add_subplot(122)
ax2 = sns.regplot(x=actual_user_score_ps4["user_score"], 
            y=actual_user_score_ps4["total_sales"], 
            line_kws={"color":"r","alpha":0.7,"lw":5})
ax2.set_title('Влияние оценок пользователей на продажи платформы PlayStation 4', size='medium', weight='bold')
ax2.set(xlabel='Оценка пользователей', ylabel='Количество глобальных продаж (млн копий)')

plt.show()
```

Посчитаем коэффициенты корреляций для платформы Sony PlayStation 4 между следующими значениями: 
1. Оценка пользователей к глобальным продажам
2. Оценка критиков к глобальным продажам

```python
#расчет корреляции
print("Коэффициент корреляции Пирсона для PS4 между оценками критиков и глобальными продажами равен:")
print(actual_critic_score_ps4['critic_score'].corr(actual_critic_score_ps4['total_sales']).round(2))
print()
print("Коэффициент корреляции Пирсона для PS4 между оценками пользователей и глобальными продажами равен:")
print(actual_user_score_ps4['user_score'].corr(actual_user_score_ps4['total_sales']).round(2))
```

Диаграмма рассеяния и рассчеты корреляции для платформы PS4 показали, что:
1) **взаимосвязь между оценками критиков и мировыми продажами игр умеренная положительная**. На диаграмме можно увидеть, что с ростом количества оценок критиков продажи игр на платформе незначильно растут. *Здесь можно говорить о том, что практически нет зависимости между значениями*.  

2) **взаимосвязь между оценками пользователей и мировыми продажами игр слабая**, практически равна нулю, то есть указывает на отсутствие линейной корреляции между двумя переменными. Это так же видно на диаграмме по линии регрессии - не смотря на увеличение оценок от пользователей, продажи игр не увеличивают. *Можем говорить о том, что никакой взаимосвязи между оценками пользователей и продажами игр нет*.

## Проверка гипотез

### Гипотеза №1

**H0**: Средние пользовательские рейтинги платформ Xbox One и PC равны.  
**H1**: Средние пользовательские рейтинги платформ Xbox One и PC разные.  

Для проверки гипотезы берем актуальные данные за последние 3 года (с 2014 по 2016 г включительно), так как, во-первых, это важно для прогноза на 2017 год, а во-вторых, платформа Xbox One была выпущена в 2013 году.

```python
#делаем срез данных: отсекаем нулевые значения оценок пользователей и выбираем платформу Xbox One
hypothesis_data_xone = actual_game_data.query('(user_score > 0.0) and (platform == "XOne")')
#вычисляем дисперсию оценок пользователей, играющих на Xbox One
statistics.variance(hypothesis_data_xone['user_score'])
```

```python
#делаем срез данных: отсекаем нулевые значения оценок пользователей и выбираем платформу PC
hypothesis_data_pc = actual_game_data.query('(user_score > 0.0) and (platform == "PC")')
#вычисляем дисперсию оценок пользователей, играющих на PC
statistics.variance(hypothesis_data_pc['user_score'])
```

Дисперсии значений оценок пользователей (`user_score`) двух выборок - для платформы XOne и для PC разные.

<div class="alert alert-success">
<font size="4", color= "seagreen"><b>✔️ Комментарий ревьюера</b></font>
    <br /> 
    <font size="3", color = "black">
<br />Важно удалить пропуски и «заглушки» перед проведением теста, молодец

```python
#определяем уровень статистической значимости
#так как набор данных у нас не гигантский, то значения 0.05 будет вполне достаточно
alpha = .05 

# проведем ttest для проверки гипотезы о равенстве среднего двух генеральных совокупностей 
# гипотеза двухсторонняя, поэтому по умолчанию встанет параметр 'two-sided'
# так как дисперсия числовых значений user_score у двух выборок разная, то укажем это в параметре equal_var
results = st.ttest_ind(
    hypothesis_data_xone['user_score'],
    hypothesis_data_pc['user_score'],
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: cредние пользовательские рейтинги платформ Xbox One и PC разные.')
else:
    print('Не получилось отвергнуть нулевую гипотезу:' 
          'cредние пользовательские рейтинги платформ Xbox One и PC одинаковые.')
```

### Гипотеза №2

**H0**: Средние пользовательские рейтинги жанров Action и Sports равны.  
**H1**: Средние пользовательские рейтинги жанров Action и Sports разные. 

Аналогично для проверки второй гипотезы берем данные за последние 3 года - с 2014 по 2016 г.

```python
#делаем срез данных: отсекаем нулевые значения оценок пользователей и выбираем жанр Action
hypothesis_data_action = actual_game_data.query('(user_score > 0.0) and (genre == "Action")')
#вычисляем дисперсию оценок пользователей для жанра Action
statistics.variance(hypothesis_data_action['user_score'])
```

```python
#делаем срез данных: отсекаем нулевые значения оценок пользователей и выбираем жанр Sports
hypothesis_data_sports = actual_game_data.query('(user_score > 0.0) and (genre == "Sports")')
#вычисляем дисперсию оценок пользователей для жанра Sports
statistics.variance(hypothesis_data_sports['user_score'])
```

Дисперсии значений оценок пользователей (`user_score`) двух выборок - игры в жанре Action и игры в жанре Sports отличаются.

```python
#определяем уровень статистической значимости
#так как набор данных у нас не гигантский, то значения 0.05 будет вполне достаточно
alpha = .05 

# проведем ttest для проверки гипотезы о равенстве среднего двух генеральных совокупностей 
# гипотеза двухсторнняя, поэтому по умолчанию встанет параметр 'two-sided'
# так как дисперсия числовых значений user_score у двух выборок разная, то укажем это в параметре equal_var
results = st.ttest_ind(
    hypothesis_data_action['user_score'],
    hypothesis_data_sports['user_score'],
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: cредние пользовательские рейтинги жанров Action и Sports разные.')
else:
    print('Не получилось отвергнуть нулевую гипотезу:' 
          'Средние пользовательские рейтинги жанров Action и Sports одинаковые.')
```

По результатам ttest-а значение p-value оказалось больше уровня статистической значимости, поэтому нулевую гипотезу не отвергаем: средние оценки пользователей приставки Xbox One и компьютера PC одинаковые.
